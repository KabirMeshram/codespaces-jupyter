```python
num_years = int(input('Enter number of years married:'))

if num_years == 1:
    print("Your first year -- great!")
if num_years == 10:
    print("Your first decade -- impressive.")
if num_years == 25:
    print("Your silver anniversary -- enjoy.")
if num_years == 50:
    print("Your golden anniversary -- impressive.")
```

```python
user_num = int(input("Enter a number:"))
div_remainder = user_num % 2
if div_remainder == 0:
    print(f'{user_num} is even.')
if div_remainder ==1:
    print(f'{user_num} is odd.')
```
