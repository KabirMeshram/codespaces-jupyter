```python
def calc_pizza_area():
    pi_val = 3.14159265

    pizza_diameter = 12.0
    pizza_radius = pizza_diameter / 2.0
    pizza_area = pi_val * pizza_radius * pizza_radius
    return pizza_area
print('12.0 inch pizza is',calc_pizza_area(),'square inches.')
```

```python
CM_PER_INCH = 2.54
INCHES_PER_FOOT = 12

def height_US_to_cm(feet, inches):
    """Converts height in feet/inches to centimeters"""
    total_inches = feet * INCHES_PER_FOOT + inches
    cm = total_inches * CM_PER_INCH
    return cm

i_feet = 6
i_inches = 4

centimeters = height_US_to_cm(i_feet, i_inches)
print('Centimeters:',centimeters)
```

```python
def split_check(amount, num_people, tax_percentage, tip_percentage):
    total_bill = amount * (1 + tax_percentage) * (1 + tip_percentage)

    per_person = total_bill / num_people

    return per_person

split_check(125.00, 2, 0.095, tip_percentage = 0.15)
```

```python
def append_to_list(value, my_list = None):
    if my_list == None:
        my_list = []

    my_list.append(value)
    return my_list

numbers = append_to_list(50)
print(numbers)
numbers = append_to_list(100)
print(numbers)
```

```python
def print_sandwich(bread, slice, *args):
    print(f'Arbitary arguments are :',*args)
    print(f'{slice} on {bread}',end=' ')
    if len(args) > 0:
        print('with', end=' ')
    for extra in args:
        print(extra, end=' ')
    print('')
print_sandwich('sourdough', 'kabab', 'mayo')
print_sandwich('wheat', 'paneer', 'mustard', 'tomato', 'lettuce')
```

```python
student_scores = [75, 84, 66, 99, 51, 65]

def get_grade_stats(scores):
    mean = sum(scores)/len(scores)

    tmp = 0
    for score in scores:
        tmp += (score - mean )**2
    std_dev = (tmp/len(scores))**0.5

    return mean, std_dev

average, standard_deviation = get_grade_stats(student_scores)

print('Average score:', average)
print('Standard deviation:', standard_deviation)
```
